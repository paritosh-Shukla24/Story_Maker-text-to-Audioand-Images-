```python
def Merge(images, waveforms):
    frame_rate = 24
    merged_data = []

    for image, waveform in zip(images, waveforms):
        waveform_duration = len(waveform) / 16000
        num_frames = int(waveform_duration * frame_rate)
        frames = [image] * num_frames
        merged_data.extend(zip(frames, [waveform] * num_frames))
    return merged_data

merged_data = Merge(image, waveforms)
```
